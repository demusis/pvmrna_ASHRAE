```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.constraints import max_norm

print(tf.__version__)
```

```python
# %load_ext tensorboard

# import datetime, os
# logs_base_dir = "./logs"
# os.makedirs(logs_base_dir, exist_ok=True)
# %tensorboard --logdir {logs_base_dir}
```

```python
# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw
# https://www.tensorflow.org/tutorials/keras/regression?hl=pt-br

def constroiModelo(n):
  modelo = keras.Sequential([
    layers.Dense(round(4*n/7) + 1, activation='relu', input_shape=[len(x_data.keys())]),
    layers.Dropout(0.2),
    layers.Dense(round(4*n/14) + 1, activation='relu'),
    layers.Dropout(0.50),
    layers.Dense(round(4*n/28) + 1, activation='relu'),
    layers.Dense(1)
  ])

  # A proposta de uma rede triangular advém de:
  #  https://ieeexplore.ieee.org/document/8540636

  # A regularização por dropout reduz possibilidade de overfitting, no caso 20% na camada de entrada e 50% na camada oculta.
  # https://arxiv.org/pdf/1207.0580.pdf
  # https://jmlr.org/papers/volume15/srivastava14a/srivastava14a.pdf

  # A justificativa do uso de ReLU:
  # https://www.sciencedirect.com/science/article/pii/B9780128188033000301
  # https://proceedings.mlr.press/v15/glorot11a/glorot11a.pdf


  otimizador = tf.keras.optimizers.Nadam(0.01)
  # Corresponde a incorporação do momento de Nesterov no algoritmo Adam
  # https://openreview.net/pdf?id=OM0jvwB8jIp57ZJjtNEZ
  
  modelo.compile(loss='mae', # Função a minimizar
                 optimizer=otimizador,
                 metrics=['mse', 'accuracy'])

  # A opção de MAE como função de perda é com a finalidade de tornar o modelo mais robusto em relação a presença de outliers.  
    
  return modelo
```

```python
import numpy as np
import time
from sklearn.model_selection import train_test_split
from tensorflow.keras import callbacks
```

```python
def processaModelo(x, y, n):
    inicio = time.time()
    modelo = constroiModelo(n)
    
    # Particiona os dados
    x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.1, shuffle= True)
    
    # CallBack
    earlystopping = callbacks.EarlyStopping(monitor ="val_loss",
                                            mode ="min", patience = 5,
                                            restore_best_weights = True)

    
    
    # modelo.summary()
    processamento = modelo.fit(x = x_treino, 
                               y = y_treino, 
                               epochs = 1500, # Máximo
                             
                               validation_data=(x_teste, y_teste), 
                               batch_size = 400,
                               verbose = 0,
                               
                               callbacks =[earlystopping] # <-
                              )
    # print("Processamendo n: ", n, ', com ', time.time()-inicio)
    return processamento
```

```python
resultados = pd.DataFrame(columns=['n', 'mae'])
```

```python
inicio = time.time()
for aux in range(93):
    df = pd.DataFrame(range(5, 300, 5), columns=['n'])
    df['mae'] = df.apply(lambda row : np.min(processaModelo(x_data, y_val, row[0]).history['val_loss']), axis = 1)

    print(aux + 1, " - Processamendo total: ", time.time()-inicio)
    resultados = resultados.append(df)
```
